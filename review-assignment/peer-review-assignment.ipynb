```python
import pandas as pd
import random

random.seed(123)

df = pd.read_csv('groups.csv')
df.head()
```

```python
d_group_id = {}
d_group_id_to_review = {}
d_reviewer = {}

for i, row in enumerate(df.iterrows()):
    group_id = row[1][0]
    title = row[1][1]
    name1 = row[1][2]
    name2 = row[1][4]
    name3 = row[1][6]
    
    d_group_id[group_id] = []
    d_group_id_to_review[group_id] = []
    
    for name in (name1, name2, name3):
        if name and isinstance(name, str):
            d_reviewer[name] = []
            d_group_id[group_id].append(name)
```

```python
#np.random.seed(123)
#group_ids = np.array(list(d_group_id.keys()))
#np.random.shuffle(group_ids)
```

```python
for name in d_reviewer:
    
    while len(d_reviewer[name]) < 2:
        
        gid = random.choice(list(d_group_id.keys()))
        if len(d_group_id_to_review[gid]) > 5:
            continue
        
        if name not in d_group_id_to_review[gid]:
            if name not in d_group_id[gid]: 
    
                # students shouldn't review each other
                skip = False
                for name2 in d_group_id_to_review[gid]:
                    if not skip:
                        for project_id_to_review in d_reviewer[name2]:
                            if name in d_group_id_to_review[project_id_to_review]:
                                #print('skipped')
                                continue
                                skip = True

                    
                d_reviewer[name].append(gid)
                d_group_id_to_review[gid].append(name)
            
```
